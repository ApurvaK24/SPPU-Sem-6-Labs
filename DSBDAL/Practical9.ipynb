```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('titanic_dataset.csv')
```

```python
def fun1(val):
    if(val=="male"):
        return 1
    else:
        return 0

df['Sex']=df['Sex'].apply(fun1)
```

```python
print(df['Sex'])
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0

df["Embarked"] = df["Embarked"].apply(fun2)
print(df['Embarked'])
```

```python
plt.figure(figsize=(10,7))
box=sns.boxplot(x='Age', y='Sex', hue='Survived', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Fare', y='Sex', hue='Survived', data=df)
plt.show()
```
