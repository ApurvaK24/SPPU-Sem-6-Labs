```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('iris_dataset.csv')
```

```python
df.head()
```

```python
# Exclude non-numeric columns from the DataFrame
numeric_df = df.select_dtypes(include='number')

# Plot the correlation heatmap
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

```python
# Step 2: List down the features and their types
# Get the column names and their data types
features_types = df.dtypes

# Print the features and their types
print("Features and their types:")
print(features_types)
```

```python
sns.boxplot(x=df['target'] ,y=df["sepal width (cm)"])
```

```python
sns.boxplot(x=df['target'] ,y=df["sepal length (cm)"])
```

```python
# Step 3: Create a histogram for each feature
# Loop through each feature and create a histogram
for column in df.columns[:-1]:  # Exclude the target column
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column], kde=True)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(df.drop(columns='target'),orient='h')
plt.title("Box plot of Features")
plt.xlabel("Value")
plt.ylabel("Feature")
plt.show()
```
